```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../Week-2_Data_Cleaning_Encoding/encoded_epl_player.csv")
df.head()
```

```python
plt.figure()
plt.hist(df['Goals'], bins=20)
plt.xlabel("Goals")
plt.ylabel("Frequency")
plt.title("Distribution of Goals")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Appearances'], df['Goals'])
plt.xlabel("Appearances")
plt.ylabel("Goals")
plt.title("Appearances vs Goals")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Age'], df['Goals'])
plt.xlabel("Age")
plt.ylabel("Goals")
plt.title("Age vs Goals")
plt.show()
```

```python
club_goals = df.groupby('Club_encoded')['Goals'].sum()

plt.figure()
club_goals.plot(kind='bar')
plt.xlabel("Club")
plt.ylabel("Total Goals")
plt.title("Total Goals by Club")
plt.show()
```

```python
position_goals = df.groupby('Position_encoded')['Goals'].mean()

plt.figure()
position_goals.plot(kind='bar')
plt.xlabel("Position")
plt.ylabel("Average Goals")
plt.title("Average Goals by Position")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Shots'], df['Goals'])
plt.xlabel("Shots")
plt.ylabel("Goals")
plt.title("Shots vs Goals")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Assists'], df['Goals'])
plt.xlabel("Assists")
plt.ylabel("Goals")
plt.title("Assists vs Goals")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Passes'], df['Assists'])
plt.xlabel("Passes")
plt.ylabel("Assists")
plt.title("Passes vs Assists")
plt.show()
```
