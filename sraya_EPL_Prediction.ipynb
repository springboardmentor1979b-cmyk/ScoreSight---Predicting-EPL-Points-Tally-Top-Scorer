```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/AI_SCORE2.csv')
```

```python
df.head()      # shows first 5 rows
df.columns     # shows all column names
df.info()      # shows data types + missing values
df.describe()  # shows summary statistics
```

```python
df.isnull().sum()
```

```python

import numpy as np

# Identify all numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Replace missing values with column mean
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

```

```python
# Count remaining missing values in numeric columns
print(df[num_cols].isna().sum())
```

```python
#Explore what each column represents

print("Preview of each column:")
for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].head())
```

```python
# Columns useful for predicting player performance
useful_cols = ['Goals', 'Assists', 'Shots', 'Accuracy', 'Appearances']

print("Useful columns for prediction:", useful_cols)

# Possible irrelevant columns
irrelevant_cols = ['Jersey Number', 'Nationality', 'Club', 'Name']

print("Irrelevant columns:", irrelevant_cols)
```

```python
#Detect zero values
print("\nZero values per column:")
print((df == 0).sum())
```

```python
#Replace zero values + missing values with mean
for col in useful_cols:
    if col in df.columns:
        # Convert 0 → NaN (if 0 is considered missing)
        df[col] = df[col].replace(0, np.nan)

        # Fill NaN with mean
        df[col] = df[col].fillna(df[col].mean())
```

```python
useful_cols = ['Goals', 'Assists', 'Shots', 'Accuracy', 'Appearances']
useful_cols = [col for col in useful_cols if col in df.columns]

print(df[useful_cols].isna().sum())

```
