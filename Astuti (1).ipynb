```python
import pandas as pd
import numpy as np

df = pd.read_csv("dataset - 2020-09-24.csv")

useful_cols = [
    'Goals', 'Assists', 'Appearances', 'Shots', 'Shots on target',
    'Shooting accuracy %', 'Passes', 'Passes per match'
]

for col in useful_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

for col in useful_cols:
    df[col] = df[col].replace(0, np.nan)
    df[col] = df[col].fillna(df[col].mean())

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["Goals"], bins=20)
plt.xlabel("Goals")
plt.ylabel("Frequency")
plt.title("Histogram of Goals")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingClassifier
from sklearn.metrics import (
    mean_absolute_error, mean_squared_error, r2_score,
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix
)
```

```python
X_reg = df[["Appearances", "Shots", "Assists", "Passes"]]
y_reg = df["Goals"]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)
```

```python
import pandas as pd
import numpy as np

# Load the original dataframe to get 'Goals' values before imputation for y_clf
original_df_for_clf = pd.read_csv("dataset - 2020-09-24.csv")
original_df_for_clf['Goals'] = pd.to_numeric(original_df_for_clf['Goals'], errors='coerce')
original_df_for_clf['Goals'] = original_df_for_clf['Goals'].fillna(0) # Handle any non-numeric or actual NaN in original 'Goals' as 0

X_clf = df[["Shots", "Passes", "Tackles"]].copy()
X_clf["Tackles"] = pd.to_numeric(X_clf["Tackles"], errors='coerce')
X_clf["Tackles"] = X_clf["Tackles"].fillna(X_clf["Tackles"].mean())

y_clf = (original_df_for_clf["Goals"] > 0).astype(int)

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)
```

```python
lr = LinearRegression()
lr.fit(X_train_reg, y_train_reg)

y_pred_lr = lr.predict(X_test_reg)
```

```python
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_clf, y_train_clf)

y_pred_log = logreg.predict(X_test_clf)
```

```python
mae = mean_absolute_error(y_test_reg, y_pred_lr)
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_lr))
r2 = r2_score(y_test_reg, y_pred_lr)

print("Linear Regression Metrics")
print("MAE:", mae)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
print("Logistic Regression Metrics")
print("Accuracy:", accuracy_score(y_test_clf, y_pred_log))
print("Precision:", precision_score(y_test_clf, y_pred_log, average="weighted"))
print("Recall:", recall_score(y_test_clf, y_pred_log, average="weighted"))
print("F1 Score:", f1_score(y_test_clf, y_pred_log, average="weighted"))
```

```python
plt.scatter(y_test_reg, y_pred_lr)
plt.xlabel("Actual Goals")
plt.ylabel("Predicted Goals")
plt.title("Actual vs Predicted Goals")
plt.show()
```

```python
cm = confusion_matrix(y_test_clf, y_pred_log)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train_reg, y_train_reg)

y_pred_rf = rf.predict(X_test_reg)
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train_clf, y_train_clf)

y_pred_gb = gb.predict(X_test_clf)
```

```python
param_grid_rf = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20]
}

grid_rf = GridSearchCV(rf, param_grid_rf, cv=3)
grid_rf.fit(X_train_reg, y_train_reg)

best_rf = grid_rf.best_estimator_
```

```python
param_grid_gb = {
    "n_estimators": [100, 200],
    "learning_rate": [0.05, 0.1]
}

grid_gb = GridSearchCV(gb, param_grid_gb, cv=3)
grid_gb.fit(X_train_clf, y_train_clf)

best_gb = grid_gb.best_estimator_
```

```python
print("Linear Regression R²:", r2_score(y_test_reg, y_pred_lr))
print("Random Forest R²:", r2_score(y_test_reg, best_rf.predict(X_test_reg)))
```

```python
print("Logistic Accuracy:", accuracy_score(y_test_clf, y_pred_log))
print("Gradient Boosting Accuracy:", accuracy_score(y_test_clf, best_gb.predict(X_test_clf)))
```
