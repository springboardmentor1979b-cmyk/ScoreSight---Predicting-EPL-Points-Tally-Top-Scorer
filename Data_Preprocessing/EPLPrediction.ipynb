Find Useful Columns And Irrelevant Columns

```python
# Useful Columns for Player Performance (Target Features):
useful_cols = [
    'Goals', 'Assists', 'Goals per match', 'Shots',
    'Shots on target', 'Shooting accuracy %',
    'Passes', 'Big chances created', 'Appearances', 'Wins', 'Losses'
]

# Irrelevant Columns (Not Needed for Performance Prediction):
irrelevant_cols = [
    'Jersey Number', 'Nationality', 'Name', 'Club'
]

print("\n Useful Columns for Performance Prediction:")
print(useful_cols)

print("\n Irrelevant Columns (likely to drop later if needed):")
print(irrelevant_cols)
```

Data Cleaning- Handle Missing Values & Handling Categorical Data

```python
import pandas as pd
import numpy as np

# Important performance columns
important_cols = ['Goals', 'Assists', 'Appearances', 'Shots', 'Shooting accuracy %']

# Replace '0' with NaN only in important columns
df[important_cols] = df[important_cols].replace(0, np.nan)

# Remove % sign from accuracy column and convert to numeric
df["Shooting accuracy %"] = (
    df["Shooting accuracy %"].astype(str).str.replace('%', '', regex=True)
)

df[important_cols] = df[important_cols].apply(pd.to_numeric, errors='coerce')

# Fill NaN (missing values) with mean of each column
df[important_cols] = df[important_cols].fillna(df[important_cols].mean())

# Output: Check if missing values still exist
print("\n Missing Values After Cleaning:")
print(df[important_cols].isna().sum())
```
