```python
import pandas as pd
import numpy as np

# -----------------------------
# 1. SORT & SAVE CLEANED DATASET
# -----------------------------
df_sorted = df.sort_values(by=df.columns[0])   # Sort by first column
df_sorted.to_csv("cleaned_sorted_dataset.csv", index=False)
print("✔ Cleaned & sorted dataset saved as cleaned_sorted_dataset.csv")


# -----------------------------
# 2. ENCODE CATEGORICAL COLUMNS
# -----------------------------
df_encoded = df.copy()

# Encode Club = Arsenal
df_encoded['Club_Arsenal'] = (df_encoded['Club'] == 'Arsenal').astype(int)

# Encode Position = Forward
df_encoded['Position_Forward'] = (df_encoded['Position'] == 'Forward').astype(int)


# Encode Home/Away = Home (SAFE CHECK)
home_col = None
possible_home_cols = ['Home/Away', 'HomeAway', 'Venue', 'Match Location', 'H/A']

for col in possible_home_cols:
    if col in df.columns:
        home_col = col
        break

if home_col:
    df_encoded['HomeAway_Home'] = (df_encoded[home_col] == 'Home').astype(int)
else:
    print("⚠ No Home/Away column found. Adding default = 0.")
    df_encoded['HomeAway_Home'] = 0


# -----------------------------
# 3. SHOW RESULTS
# -----------------------------
print("\n✔ Encoded Columns Preview:")
print(df_encoded[['Club_Arsenal', 'Position_Forward', 'HomeAway_Home']].head())

# Save encoded dataset
df_encoded.to_csv("encoded_dataset.csv", index=False)
print("\n✔ Encoded dataset saved as encoded_dataset.csv")
```

# Task-3: Exploratory data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# =========================================
# 1. Load Dataset
# =========================================
file_path = "/content/dataset - 2020-09-24.csv"   # <-- update if needed
df = pd.read_csv(file_path)
print("Dataset Loaded Successfully!")
print(df.head())

# =========================================
# 2. Clean Percentage Column ("Tackle success %")
# =========================================

if "Tackle success %" in df.columns:

    df["Tackle success %"] = (
        df["Tackle success %"]
        .astype(str)
        .str.replace("%", "", regex=False)
        .str.strip()
    )

    df["Tackle success %"] = pd.to_numeric(df["Tackle success %"], errors="coerce")

    print("\nCleaned 'Tackle success %' column:")
    print(df["Tackle success %"].head())

else:
    print("\nColumn 'Tackle success %' not found.")


# ====================================================
# 3. Exploratory Data Analysis (All Graphs)
# ====================================================

# -----------------------------
# Histogram: Goals
# -----------------------------
plt.figure(figsize=(6,4))
plt.hist(df["Goals"].dropna())
plt.xlabel("Goals")
plt.ylabel("Frequency")
plt.title("Histogram of Goals")
plt.show()


# -----------------------------
# Scatter: Appearances vs Goals
# -----------------------------
plt.figure(figsize=(6,4))
plt.scatter(df["Appearances"], df["Goals"])
plt.xlabel("Appearances")
plt.ylabel("Goals")
plt.title("Appearances vs Goals")
plt.show()


# -----------------------------
# Scatter: Age vs Goals
# -----------------------------
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["Goals"])
plt.xlabel("Age")
plt.ylabel("Goals")
plt.title("Age vs Goals")
plt.show()


# -----------------------------
# Bar Chart: Club vs Total Goals
# -----------------------------
plt.figure(figsize=(10,5))
club_goals = df.groupby("Club")["Goals"].sum()
plt.bar(club_goals.index, club_goals.values)
plt.xticks(rotation=90)
plt.xlabel("Club")
plt.ylabel("Total Goals")
plt.title("Total Goals by Club")
plt.show()


# -----------------------------
# Bar Chart: Position vs Avg Goals
# -----------------------------
plt.figure(figsize=(6,4))
pos_goals_avg = df.groupby("Position")["Goals"].mean()
plt.bar(pos_goals_avg.index, pos_goals_avg.values)
plt.xlabel("Position")
plt.ylabel("Average Goals")
plt.title("Average Goals by Position")
plt.show()


# -----------------------------
# Scatter: Shots vs Goals
# -----------------------------
plt.figure(figsize=(6,4))
plt.scatter(df["Shots"], df["Goals"])
plt.xlabel("Shots")
plt.ylabel("Goals")
plt.title("Shots vs Goals")
plt.show()


# -----------------------------
# Scatter: Assists vs Goals
# -----------------------------
plt.figure(figsize=(6,4))
plt.scatter(df["Assists"], df["Goals"])
plt.xlabel("Assists")
plt.ylabel("Goals")
plt.title("Assists vs Goals")
plt.show()


# -----------------------------
# Scatter: Passes vs Assists
# -----------------------------
plt.figure(figsize=(6,4))
plt.scatter(df["Passes"], df["Assists"])
plt.xlabel("Passes")
plt.ylabel("Assists")
plt.title("Passes vs Assists")
plt.show()


# -----------------------------
# Scatter: Tackles vs Tackle Success %
# -----------------------------
plt.figure(figsize=(6,4))
plt.scatter(df["Tackles"], df["Tackle success %"])
plt.xlabel("Tackles")
plt.ylabel("Tackle Success %")
plt.title("Tackles vs Tackle Success %")
plt.show()


# -----------------------------
# Bar Chart: Nationality vs Total Goals
# -----------------------------
plt.figure(figsize=(10,5))
nat_goals = df.groupby("Nationality")["Goals"].sum().sort_values(ascending=False)
plt.bar(nat_goals.index, nat_goals.values)
plt.xticks(rotation=90)
plt.xlabel("Nationality")
plt.ylabel("Total Goals")
plt.title("Total Goals by Nationality")
plt.show()
```

# Task-4: Feature Scaling

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# =====================================
# Load dataset
# =====================================
file_path = "/content/dataset - 2020-09-24.csv"
df = pd.read_csv(file_path)

print("Dataset Loaded Successfully!")


# =======================================================
# STEP 1 — PRINT RAW VALUES (Goals, Appearances, Passes)
# =======================================================
print("\n===== RAW VALUES =====")
print(df[["Goals", "Appearances", "Passes"]].head())


# =======================================================
# STEP 2 — MIN-MAX SCALING
# Columns: Goals, Appearances, Passes, Assists
# =======================================================
minmax_cols = ["Goals", "Appearances", "Passes", "Assists"]

scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[minmax_cols] = scaler_minmax.fit_transform(df[minmax_cols])

print("\n===== MIN-MAX SCALED (0–1) =====")
print(df_minmax[minmax_cols].head())


# =======================================================
# STEP 3 — STANDARDIZATION (Z-SCORE)
# Columns: Goals, Shots, Passes
# =======================================================
zscore_cols = ["Goals", "Shots", "Passes"]

scaler_std = StandardScaler()
df_zscore = df.copy()
df_zscore[zscore_cols] = scaler_std.fit_transform(df[zscore_cols])

print("\n===== Z-SCORE STANDARDIZED =====")
print(df_zscore[zscore_cols].head())


# =======================================================
# STEP 4 — SCATTER PLOTS BEFORE VS AFTER SCALING
# Example: Appearances vs Goals
# =======================================================

# ----------- RAW SCATTER -------------
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.scatter(df["Appearances"], df["Goals"])
plt.xlabel("Appearances")
plt.ylabel("Goals")
plt.title("Before Scaling: Appearances vs Goals")

# ----------- MIN-MAX SCALED SCATTER -------------
plt.subplot(1,2,2)
plt.scatter(df_minmax["Appearances"], df_minmax["Goals"])
plt.xlabel("Appearances (Scaled)")
plt.ylabel("Goals (Scaled)")
plt.title("After Min-Max Scaling")

plt.show()


# -------------------------------------------------------
# Scatter before vs after (Z-SCORE)
# Example: Shots vs Goals
# -------------------------------------------------------
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.scatter(df["Shots"], df["Goals"])
plt.xlabel("Shots")
plt.ylabel("Goals")
plt.title("Before Scaling: Shots vs Goals")

plt.subplot(1,2,2)
plt.scatter(df_zscore["Shots"], df_zscore["Goals"])
plt.xlabel("Shots (Z-score)")
plt.ylabel("Goals (Z-score)")
plt.title("After Standardization")

plt.show()
```
