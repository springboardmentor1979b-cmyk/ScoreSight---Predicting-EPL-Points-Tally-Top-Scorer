Removing Irrelevant Columns

```python
irrelevant = ['Jersey Number']  # example
for col in irrelevant:
    if col in df.columns:
        print(col, "can be dropped")
```

Identifying Missing Values

```python
df.isna().sum()
```

Handling Zero & Missing Values

```python
important_cols = ['Goals', 'Assists', 'Shots', 'Appearances']

for col in important_cols:
    print(col, (df[col] == 0).sum(), "zeros")
```

```python
import numpy as np

for col in important_cols:
    df[col] = df[col].replace(0, np.nan)
    df[col] = df[col].fillna(df[col].mean())
```

```python
df.isna().sum()
df.describe()
```

Dataset after sorting and encoding

```python
# -----------------------------
# 1) SORT the cleaned dataset
# -----------------------------
df = df.sort_values(by='Goals', ascending=False)   # you can change to 'Age' or 'Assists' if mam told specific

# --------------------------------------------
# 2) ENCODE categorical values as instructed
#    Club → Arsenal = 1 else 0
#    Position → Forward = 1 else 0
#    Home/Away → Home = 1 else 0
# --------------------------------------------

# If dataset doesn't have Home/Away column, create one
if 'Home/Away' not in df.columns:
    df['Home/Away'] = 'Home'   # You may change later if needed

df['Club_Encoded'] = df['Club'].apply(lambda x: 1 if x=='Arsenal' else 0)
df['Position_Encoded'] = df['Position'].apply(lambda x: 1 if x=='Forward' else 0)
df['Home_Away_Encoded'] = df['Home/Away'].apply(lambda x: 1 if x=='Home' else 0)

# --------------------------------------------
# 3) SAVE dataset after sorting + encoding
# --------------------------------------------
output_path = '/content/drive/MyDrive/sorted_cleaned_encoded_dataset.csv'
df.to_csv(output_path, index=False)

print("Task completed")
print("Saved file at:", output_path)
df.head()
```

Replace all NaN values with column mean

```python
import numpy as np

# Select only numeric columns for NA replacement
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Replace NaN with column mean
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

print("All NaN values in numeric columns replaced with mean.")

# Save cleaned dataset again after removing NaN
output_path = '/content/drive/MyDrive/sorted_cleaned_encoded_dataset_noNaN.csv'
df.to_csv(output_path, index=False)

print(" File saved without NaN at:", output_path)

df.head()
```

🔹 TRAIN – TEST SPLIT

```python
from sklearn.model_selection import train_test_split

# ---------- REGRESSION DATA ----------
X_reg = df[['Appearances','Shots','Passes','Assists','Age']]
y_reg = df['Goals']

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

# ---------- CLASSIFICATION DATA ----------
# Create match outcome column
df['Match_Result'] = df['Wins'] - df['Losses']
df['Match_Result'] = df['Match_Result'].apply(lambda x: 1 if x>0 else (0 if x==0 else -1))

X_clf = df[['Goals','Assists','Passes','Shots']]
y_clf = df['Match_Result']

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)
```

BASELINE MODELS

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train_reg, y_train_reg)

y_pred_reg = lr.predict(X_test_reg)
```

Logistic Regression (Match Outcome)

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_clf, y_train_clf)

y_pred_clf = log_reg.predict(X_test_clf)
```

MODEL EVALUATION

```python
#Regression Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
r2 = r2_score(y_test_reg, y_pred_reg)

print("Regression Metrics")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2:", r2)
```

Classification Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Classification Metrics")
print("Accuracy:", accuracy_score(y_test_clf, y_pred_clf))
print("Precision:", precision_score(y_test_clf, y_pred_clf, average='weighted'))
print("Recall:", recall_score(y_test_clf, y_pred_clf, average='weighted'))
print("F1:", f1_score(y_test_clf, y_pred_clf, average='weighted'))
```

VISUALIZATION

ADVANCED MODELS

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(X_train_reg, y_train_reg)

rf_pred = rf.predict(X_test_reg)

print("Random Forest R2:", r2_score(y_test_reg, rf_pred))
```

Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(X_train_clf, y_train_clf)

gb_pred = gb.predict(X_test_clf)

print("GB Accuracy:", accuracy_score(y_test_clf, gb_pred))
```

HYPERPARAMETER TUNING (GridSearchCV)

```python
#Random Forest (Regression)
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators':[50,100],
    'max_depth':[None,10,20]
}

grid = GridSearchCV(RandomForestRegressor(random_state=42),
                    param_grid,
                    cv=3,
                    scoring='r2')

grid.fit(X_train_reg, y_train_reg)

best_rf = grid.best_estimator_
best_pred = best_rf.predict(X_test_reg)

print("Best RF R2:", r2_score(y_test_reg, best_pred))
```

PERFORMANCE COMPARISON

```python
print("MODEL COMPARISON")
print("Linear Regression R2:", r2)
print("Random Forest R2:", r2_score(y_test_reg, rf_pred))
print("Tuned Random Forest R2:", r2_score(y_test_reg, best_pred))

print("Logistic Accuracy:", accuracy_score(y_test_clf, y_pred_clf))
print("Gradient Boosting Accuracy:", accuracy_score(y_test_clf, gb_pred))
```
