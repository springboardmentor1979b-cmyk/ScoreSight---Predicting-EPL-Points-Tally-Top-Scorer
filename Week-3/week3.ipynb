```python
import pandas as pd

df = pd.read_csv(
    "/content/drive/MyDrive/infosys internship/dataset - 2020-09-24.csv"
)

df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['Goals'], bins=20)
plt.xlabel("Goals")
plt.ylabel("Frequency")
plt.title("Distribution of Goals")
plt.show()
```

```python

plt.figure()
plt.scatter(df['Appearances'], df['Goals'])
plt.xlabel("Appearances")
plt.ylabel("Goals")
plt.title("Appearances vs Goals")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Age'], df['Goals'])
plt.xlabel("Age")
plt.ylabel("Goals")
plt.title("Age vs Goals")
plt.show()
```

```python
club_goals = df.groupby('Club')['Goals'].sum()

plt.figure(figsize=(10,5))
club_goals.plot(kind='bar')
plt.xlabel("Club")
plt.ylabel("Total Goals")
plt.title("Total Goals by Club")
plt.show()
```

```python
avg_goals_position = df.groupby('Position')['Goals'].mean()

avg_goals_position
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
avg_goals_position.plot(kind='bar')
plt.xlabel("Position")
plt.ylabel("Average Goals")
plt.title("Average Goals by Player Position")
plt.xticks(rotation=0)
plt.show()
```
