```python
irrelevant = ['Jersey Number', 'Nationality']  # example
for col in irrelevant:
    if col in df.columns:
        print(col, "can be dropped")
```

```python
df = df.drop(columns=irrelevant, errors='ignore')
```

```python
df.isna().sum()
```

```python
important_cols = ['Goals', 'Assists', 'Shots', 'Appearances']

for col in important_cols:
    print(col, (df[col] == 0).sum(), "zeros")
```

```python
import numpy as np

for col in important_cols:
    df[col] = df[col].replace(0, np.nan)
    df[col] = df[col].fillna(df[col].mean())
```
