```python
#load the csv
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/infosys internship/dataset - 2020-09-24.csv")
df.head()

from google.colab import files
uploaded = files.upload()  # Upload rf_regression_model.pkl and gb_classification_model.pkl
import joblib

rf_model = joblib.load("rf_regression_model.pkl")
gb_model = joblib.load("gb_classification_model.pkl")

```

Make predictions on a sample of the new dataset

```python
sample = df.sample(5)  # Pick random 5 rows

# Impute missing values with the mean of the respective columns from the original DataFrame
for col in ['Appearances', 'Shots', 'Passes', 'Assists']:
    if col in sample.columns and sample[col].isnull().any():
        mean_val = df[col].mean() # Use mean from original df
        sample[col].fillna(mean_val, inplace=True)

# Predict Goals (Regression)
sample['Predicted_Goals'] = rf_model.predict(
    sample[['Appearances', 'Shots', 'Passes', 'Assists']]
)

# Predict Match Result (Classification)
sample['Predicted_Result'] = gb_model.predict(
    sample[['Shots', 'Passes', 'Assists']]
)

sample
```
