Handling Missing Values in Numeric Columns

```python

import numpy as np

# Identify all numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Replace missing values with column mean
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

```

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingClassifier
from sklearn.metrics import (
    mean_absolute_error, mean_squared_error, r2_score,
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, ConfusionMatrixDisplay
)
```

```python
# _______________________________________________
# 1) TRAIN AND TEST SPLIT
# Divide dataset into training and testing sets
# ______________________________________________

# Create target variable for classification (Match Outcome)
# 2 = Win, 1 = Draw, 0 = Loss
def match_outcome(row):
    if row['Wins'] > row['Losses']:
        return 2
    elif row['Wins'] == row['Losses']:
        return 1
    else:
        return 0

df['Match_Outcome'] = df.apply(match_outcome, axis=1)

# Regression Data (Player Performance Prediction)
X_reg = df[['Appearances', 'Shots', 'Assists', 'Passes']]
y_reg = df['Goals']

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

#  Classification Data (Match Outcome Prediction)
X_clf = df[['Goals', 'Assists', 'Shots', 'Passes']]
y_clf = df['Match_Outcome']

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)
```

```python
# =_________________________________________
# 2) BASELINE MODELS
# Linear Regression (Regression)
# Logistic Regression (Classification)
# ___________________________________________

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import StandardScaler

# LINEAR REGRESSION (Player Performance)
lr = LinearRegression()
lr.fit(X_train_reg, y_train_reg)

# Predictions
y_pred_lr = lr.predict(X_test_reg)


#LOGISTIC REGRESSION (Match Outcome)
# Scale features to avoid convergence warning
scaler = StandardScaler()

X_train_clf_scaled = scaler.fit_transform(X_train_clf)
X_test_clf_scaled = scaler.transform(X_test_clf)

log_reg = LogisticRegression(max_iter=5000)
log_reg.fit(X_train_clf_scaled, y_train_clf)

# Predictions
y_pred_log = log_reg.predict(X_test_clf_scaled)

```

```python
#__________________________________________________
# 3) MODEL EVALUATION
# Regression: MAE, RMSE, R²
# Classification: Accuracy, Precision, Recall, F1
#__________________________________________________

# Regression Evaluation
print("Linear Regression Evaluation")
print("MAE:", mean_absolute_error(y_test_reg, y_pred_lr))
print("RMSE:", np.sqrt(mean_squared_error(y_test_reg, y_pred_lr)))
print("R² Score:", r2_score(y_test_reg, y_pred_lr))

# Classification Evaluation
print("\nLogistic Regression Evaluation")
print("Accuracy:", accuracy_score(y_test_clf, y_pred_log))
print("Precision:", precision_score(y_test_clf, y_pred_log, average='weighted'))
print("Recall:", recall_score(y_test_clf, y_pred_log, average='weighted'))
print("F1 Score:", f1_score(y_test_clf, y_pred_log, average='weighted'))
```

```python
#___________________________________________________
# 4) VISUALIZATION TASKS
# Loss curves, Actual vs Predicted, Confusion Matrix
#____________________________________________________

# Actual vs Predicted (Regression)
plt.scatter(y_test_reg, y_pred_lr)
plt.xlabel("Actual Goals")
plt.ylabel("Predicted Goals")
plt.title("Actual vs Predicted Goals (Linear Regression)")
plt.show()

#Confusion Matrix (Classification)
cm = confusion_matrix(y_test_clf, y_pred_log)
disp = ConfusionMatrixDisplay(cm, display_labels=["Loss", "Draw", "Win"])
disp.plot()
plt.title("Confusion Matrix – Match Outcomes")
plt.show()
```

```python
#________________________________________
# 5) ADVANCED MODELS
# Random Forest Regressor
# Gradient Boosting Classifier
#_______________________________________


#Random Forest Regressor
rf_reg = RandomForestRegressor(random_state=42)
rf_reg.fit(X_train_reg, y_train_reg)
y_pred_rf = rf_reg.predict(X_test_reg)

print("Random Forest R²:", r2_score(y_test_reg, y_pred_rf))

#Gradient Boosting Classifier
gb_clf = GradientBoostingClassifier()
gb_clf.fit(X_train_clf, y_train_clf)
y_pred_gb = gb_clf.predict(X_test_clf)

print("Gradient Boosting Accuracy:", accuracy_score(y_test_clf, y_pred_gb))
```

```python
#__________________________________________
# 6) HYPERPARAMETER TUNING
# GridSearchCV and RandomizedSearchCV
#__________________________________________

# GridSearchCV for Random Forest
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20]
}

grid_search_rf = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid,
    cv=3,
    scoring='r2'
)

grid_search_rf.fit(X_train_reg, y_train_reg)
best_rf = grid_search_rf.best_estimator_

print("Best Random Forest R²:", r2_score(y_test_reg, best_rf.predict(X_test_reg)))

#  RandomizedSearchCV for Gradient Boosting
param_dist = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 0.2]
}

random_search_gb = RandomizedSearchCV(
    GradientBoostingClassifier(),
    param_dist,
    n_iter=5,
    cv=3
)

random_search_gb.fit(X_train_clf, y_train_clf)
best_gb = random_search_gb.best_estimator_

print("Best Gradient Boosting Accuracy:",
      accuracy_score(y_test_clf, best_gb.predict(X_test_clf)))
```

```python

# 7) PERFORMANCE COMPARISON

models_reg = ['Linear Regression', 'Random Forest', 'Tuned Random Forest']
r2_scores = [
    r2_score(y_test_reg, y_pred_lr),
    r2_score(y_test_reg, y_pred_rf),
    r2_score(y_test_reg, best_rf.predict(X_test_reg))
]

plt.bar(models_reg, r2_scores)
plt.xlabel("Regression Models")
plt.ylabel("R2 Score")
plt.title("Regression Model Performance Comparison")
plt.show()

models_clf = ['Logistic Regression', 'Gradient Boosting', 'Tuned Gradient Boosting']
accuracy_scores = [
    accuracy_score(y_test_clf, y_pred_log),
    accuracy_score(y_test_clf, y_pred_gb),
    accuracy_score(y_test_clf, best_gb.predict(X_test_clf))
]

plt.bar(models_clf, accuracy_scores)
plt.xlabel("Classification Models")
plt.ylabel("Accuracy")
plt.title("Classification Model Performance Comparison")
plt.show()
```
