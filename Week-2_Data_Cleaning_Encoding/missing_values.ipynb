```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv("epl_players.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
important_cols = ['Goals', 'Assists', 'Appearances', 'Shots', 'Passes']

for col in important_cols:
    mean_val = df[col].replace(0, np.nan).mean()  # consider 0 as missing
    df[col] = df[col].replace(0, np.nan)
    df[col].fillna(mean_val, inplace=True)
```

```python
df[important_cols].isnull().sum()
```

```python
df.to_csv("epl_players_cleaned.csv", index=False)
print("Cleaned dataset saved successfully!")
```
