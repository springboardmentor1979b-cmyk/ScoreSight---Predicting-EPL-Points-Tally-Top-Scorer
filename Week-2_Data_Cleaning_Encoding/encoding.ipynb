```python
import pandas as pd

df = pd.read_csv("epl_players_cleaned.csv")
df.head()

```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['Club_encoded'] = encoder.fit_transform(df['Club'])
df['Position_encoded'] = encoder.fit_transform(df['Position'])
```

```python
df.drop(['Club', 'Position'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.to_csv("encoded_epl_player.csv", index=False)
```
