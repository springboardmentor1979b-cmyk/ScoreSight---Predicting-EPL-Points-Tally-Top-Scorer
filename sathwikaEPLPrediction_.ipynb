## Column Categorization (Relevant/Irrelevant)



```python
relevant_columns = [
    'Age', 'Appearances', 'Wins', 'Losses', 'Goals', 'Goals per match',
    'Headed goals', 'Goals with right foot', 'Goals with left foot',
    'Penalties scored', 'Freekicks scored', 'Shots', 'Shots on target',
    'Shooting accuracy %', 'Hit woodwork', 'Big chances missed',
    'Clean sheets', 'Goals conceded', 'Tackles', 'Tackle success %',
    'Last man tackles', 'Blocked shots', 'Interceptions', 'Clearances',
    'Headed Clearance', 'Clearances off line', 'Recoveries', 'Duels won',
    'Duels lost', 'Successful 50/50s', 'Aerial battles won',
    'Aerial battles lost', 'Own goals', 'Errors leading to goal', 'Assists',
    'Passes', 'Passes per match', 'Big chances created', 'Crosses',
    'Cross accuracy %', 'Through balls', 'Accurate long balls', 'Saves',
    'Penalties saved', 'Punches', 'High Claims', 'Catches',
    'Sweeper clearances', 'Throw outs', 'Goal Kicks', 'Yellow cards',
    'Red cards', 'Fouls', 'Offsides'
]

irrelevant_columns = [
    'Name', 'Jersey Number', 'Club', 'Position', 'Nationality'
]

print("Relevant Columns:", relevant_columns)
print("\nIrrelevant Columns:", irrelevant_columns)
```

## Missing Value Identification




```python
print("Total NaN values per column:\n", df.isnull().sum())
```

```python
print("\nTotal '0' values per relevant column (which might indicate missing data for numerical columns):")
for col in relevant_columns:
    if df[col].dtype in ['int64', 'float64']:
        zero_count = (df[col] == 0).sum()
        if zero_count > 0:
            print(f"  {col}: {zero_count}")
```

## Missing Value Imputation



```python
for col in relevant_columns:
    # Only process numerical columns
    if df[col].dtype in ['int64', 'float64']:
        # Create a temporary series where '0's are treated as NaN for mean calculation
        temp_series = df[col].replace(0, np.nan)

        # Calculate the mean, ignoring both original NaNs and '0's (now NaNs)
        col_mean = temp_series.mean()

        # Impute original NaN values with the calculated mean
        df[col] = df[col].fillna(col_mean)

        # Impute '0' values with the calculated mean (only if not already filled as NaN)
        df[col] = df[col].replace(0, col_mean)

print("Missing values (NaNs and '0's) in relevant numerical columns have been imputed with their respective column means.")
```

## Verification of Changes


```python
print("Total NaN values per column after imputation:\n", df.isnull().sum())
```

```python
print(df.info())
```

## Clean and Convert Percentage Columns



```python
percentage_columns = ['Shooting accuracy %', 'Tackle success %', 'Cross accuracy %']

for col in percentage_columns:
    # Remove '%' sign and convert to float
    df[col] = df[col].astype(str).str.replace('%', '').astype(float)

    # Impute missing values (NaN and 0) with the mean of non-zero, non-NaN values
    # Create a temporary series where '0's are treated as NaN for mean calculation
    temp_series = df[col].replace(0, np.nan)

    # Calculate the mean, ignoring both original NaNs and '0's (now NaNs)
    col_mean = temp_series.mean()

    # Impute original NaN values with the calculated mean
    df[col] = df[col].fillna(col_mean)

    # Impute '0' values with the calculated mean (only if not already filled as NaN)
    df[col] = df[col].replace(0, col_mean)

print("Percentage columns have been cleaned, converted to float, and imputed.")
```

```python
print(df.info())
print("\nTotal NaN values per column after percentage column imputation:\n", df[['Shooting accuracy %', 'Tackle success %', 'Cross accuracy %']].isnull().sum())
```

## Handle Remaining Missing Values



```python
jersey_number_mode = df['Jersey Number'].mode()[0]
df['Jersey Number'].fillna(jersey_number_mode, inplace=True)

nationality_mode = df['Nationality'].mode()[0]
df['Nationality'].fillna(nationality_mode, inplace=True)

print("Missing values in 'Jersey Number' and 'Nationality' columns have been imputed with their respective modes.")
```

```python
jersey_number_mode = df['Jersey Number'].mode()[0]
df['Jersey Number'] = df['Jersey Number'].fillna(jersey_number_mode)

nationality_mode = df['Nationality'].mode()[0]
df['Nationality'] = df['Nationality'].fillna(nationality_mode)

print("Missing values in 'Jersey Number' and 'Nationality' columns have been imputed with their respective modes.")
```

```python
print("Total NaN values after 'Jersey Number' and 'Nationality' imputation:\n", df[['Jersey Number', 'Nationality']].isnull().sum())
```

## Encode Specific Categorical Features


```python
df['Club_Arsenal'] = (df['Club'] == 'Arsenal').astype(int)
df['Position_Forward'] = (df['Position'] == 'Forward').astype(int)

print("New binary columns 'Club_Arsenal' and 'Position_Forward' have been created.")
```
