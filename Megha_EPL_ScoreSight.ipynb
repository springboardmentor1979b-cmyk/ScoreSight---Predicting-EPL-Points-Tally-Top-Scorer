#List of Irrelevant columns

```python
irrelevant=['Name','Jersey Number','Nationality','Tackles','Tackle success %','Last man tackles','Blocked shots','Interceptions',
'Clearances', 'Headed Clearance','Clearances off line','Recoveries','Duels won','Duels lost', 'Successful 50/50s', 'Aerial battles won', 'Aerial battles lost','Own goals', 'Errors leading to goal',
'Saves','Penalties saved','Punches', 'High Claims', 'Catches','Sweeper clearances','Throw outs','Goal Kicks','Yellow cards','Red cards','Fouls','Offsides'
]
df_relevant=df.drop(columns=irrelevant, axis=1)
df_relevant.columns
```

```python
df_relevant.isna().value_counts()
```

```python
df_relevant.isnull().sum()
```

##Removing null values and replacing it with mean

```python
numerical = df_relevant.select_dtypes(include=['int64', 'float64']).columns
categorical= df_relevant.select_dtypes(include=['object']).columns

df_relevant[numerical] = df_relevant[numerical].mask(df_relevant == 0).fillna(df_relevant[numerical].mean())
df_relevant[categorical] = df_relevant[categorical].mask(df_relevant == 0).fillna(df_relevant[categorical].mode().iloc[0])

df_relevant.isnull().sum()
```

```python
df_relevant.head()
```

#Encode Categorical variables

```python
categorical_cols = ['Club', 'Position']
df_encoded = pd.get_dummies(df_relevant, columns=categorical_cols, drop_first=True, dtype=int)
```
